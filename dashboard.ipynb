```python
#!/usr/bin/pyth
import PIL
from PIL import Image, ImageTk 
from tkinter import *
from tkinter import ttk

brief_list = [] 
model_list = []
inputs_list =[]
name_list =  []
test_data_list = []
# current_inps_lbs = []
main = Tk()
main.title('Dashboard')
main.geometry('1000x1000')
mid = 0 

rows = 0
while rows < 100:
    main.rowconfigure(rows, weight=1)
    main.columnconfigure(rows, weight=1)
    rows += 1

##Main Frame

notebook = ttk.Notebook(main)
demo = Frame(notebook)
arc = Frame(notebook)
notebook.add(demo, text='Demo')
notebook.add(arc, text='Architecture')
notebook.grid(row=1, column=0, columnspan=100, rowspan=50, sticky='NESW')

#model importing functions
def modelimport(model,name,quote,inputs,testdata):
    brief_list.append(quote)
    model_list.append(model)
    name_list.append(name)
    test_data_list.append(testdata)
    menubutton(lf,name,model)
    inputs_list.append(inputs)
    return


def mclick(x):
    global mid
    mindex = model_list.index(x)
    mid = mindex
    T.config(text = brief_list[mindex])
    mfh.config(text = name_list[mindex])
#     tdata_display(test_data_list[mindex])
    tdata.delete('1.0', END)
    tdata.insert(END, 'Test Data\n' + str(test_data_list[mindex]))
    inputs_display(inputs_list[mindex])
    return


cinputs ={}
cinplbls =[]
def inputs_display(li):
#     rf.pack_forget()
#     rf.destroy()
    clear_inps()
    for i in li:
        lbl,ent = inputent(rf,i)
        cinputs[i[0]] = ent
        cinplbls.append(lbl)
    return

#running the test
def runtest(mid):
    return

#result view
def bfdisplay(data,imgs,root):
    for i in imgs:
        bimg = Label(root, image = i)
        bimg.image = i
        bimg.pack(side=LEFT)
    rtdata = Text(root, height=12, width=25)
    rtdata.insert(END, data)
    rtdata.pack(side=LEFT)
    return


#UI element Functions
def fhead(frame,tex):
    heading = Label(frame,text=tex, font='Helvetica 12 bold')
    heading.pack(fill='x')
    return heading

def menubutton(frame,tex,model):
    mbu = Button(frame, text=tex,command=lambda :mclick(model))
    mbu.pack(fill='x')
    return mbu





def tdata_display(frame):
#     S = Scrollbar(mf)

    tdata = Text(frame, height=12, width=30)
#     tdata.insert(END,'\nTest Data\n', 'big')
    tdata.insert(END, 'Select any model from menu')
    tdata.pack(fill='x')
#     S.config(command=testdata.yview)
#     testdata.config(yscrollcommand=S.set)
    return tdata





def inputent(frame,inp):
    lbl = Label(frame,text=inp[0])
    lbl.pack()
    if len(inp)<2:  
        ent = Entry(frame)
        ent.pack(side=TOP,fill='x')
    else:
        ent = combox(frame,inp[1])
        ent.pack(side=TOP,fill='x')
    return lbl,ent
def combox(window,li):
    combo = ttk.Combobox(window)
    combo['values']= list(li)
    combo.current(1) #set the selected item
    return combo
def clear_inps():
    for i in list(cinputs.keys()):
        cinputs[i].destroy()
    for i in cinplbls:
        i.destroy()
#     mbu.destroy()
    return





###helping functions
def resimg(path):
    image = PIL.Image.open(path)
    photo = ImageTk.PhotoImage(image)
    return photo






def close():
    main.mainloop()
    return

###demo




#left frame
lf = Frame(demo, highlightbackground="black", highlightcolor="black",width=240, height=500, highlightthickness=1,bd= 0)
lf.grid(row=0, column=0, rowspan=30,columnspan=20,sticky="nsew")
lfh = fhead(lf,'Menu')
# but1 = menubutton(lf,'Option1',model)







#middle Frame
mf = Frame(demo, highlightbackground="black", highlightcolor="black",width=520, height=500, highlightthickness=1,bd= 0)
mf.grid(row=0, column=20,rowspan=30,columnspan=60,sticky="nsew")
mfh = fhead(mf,'Controls')
T = Label(mf,text="Select any model from menu")
T.pack(fill='x')
tdata = tdata_display(mf)
# mbu = Button(mf, text="fhm",command=quit)
# mbu.pack(side=TOP)



runtestlbl = Label(mf,text="Run the test with the test data displayed above as inputs")
runtestlbl.pack(side = BOTTOM)
testbutton = Button(mf, text="Run Test",command=runtest(mid))
testbutton.pack(side=BOTTOM)



#right Frame
rf = Frame(demo, highlightbackground="black", highlightcolor="black",width=240, height=500, highlightthickness=1,bd= 0)
rf.grid(row=0, column=80 ,rowspan=30,columnspan=20,sticky="nsew")
rfh = fhead(rf,'Custom Inputs')
# rlbl,rinp = inputent(rf,['test',[1,2,3,4]])
# rlbl1,rip1 = inputent(rf,['another'])



cus_runtestlbl = Label(rf,text="Predict with the custom inputs given above")
cus_runtestlbl.pack(side = BOTTOM)
cus_testbutton = Button(rf, text="Run test",command=runtest(mid))
cus_testbutton.pack(side=BOTTOM)

#stabilizing frame
sft = Frame(demo, highlightbackground="black", highlightcolor="black", height=500, highlightthickness=1,bd= 0)
sft.grid(row=0, column=100, rowspan=30,columnspan=1,sticky="nsew")
sfb = Frame(demo, highlightbackground="black", highlightcolor="black", height=500, highlightthickness=1,bd= 0)
sfb.grid(row=30, column=100, rowspan=20,columnspan=1,sticky="nsew")
sfbv = Frame(demo, highlightbackground="black", highlightcolor="black",width=1000, highlightthickness=1,bd= 0)
sfbv.grid(row=30, column=0,columnspan=100,sticky="nsew")


#Results Frame
bf = Frame(demo, highlightbackground="black", highlightcolor="black",width=1000, height=300, highlightthickness=1,bd= 0)
bf.grid(row=30, column=0,rowspan=20,columnspan=100,sticky="nsew")
bfh = fhead(bf,'Results')






#arc
inputs = [['test'],['teq',[1,2,3]],['another']]
inputs1 = [['anottt'],['teq',[1,2,3]],['another',['tets','group']]]
testdata=[[1,2,3],[4,5,6]]
testdata1 = [[1,1,1],[3.4,5,7]]
img1 = ImageTk.PhotoImage(PIL.Image.open("barg.png").resize((400, 300), PIL.Image.ANTIALIAS))
img2 = ImageTk.PhotoImage(PIL.Image.open("test.png").resize((400, 300), PIL.Image.ANTIALIAS))


modelimport(model,'Model1',"Model with architecture as 64-32-2-1",inputs,testdata)
modelimport(model1,'Model2',"Model with architecture as 64-64-1",inputs1,testdata1)



# import matplotlib.pyplot as plt
# a =[]
# b=[]
# matimg = plt.plot(a)
# test_img = resimg('test.png')
# clear_inps()


bfdisplay('test data on bottom frame for demo',[img1,img2],bf)


# bimg = Label(bf, image = img1)
# bimg.pack()
# res_img_display = Label(bf,image = test_img)
    

scrollbar = Scrollbar(bf)
scrollbar.pack(side = RIGHT, fill = Y)
 
# area = Text(bf, yscrollcommand = scrollbar.set,  )
# area.pack(expand=True, fill='both')
 
# scrollbar.config(command = area.yview)
    

close()
```

```python
#!/usr/bin/pyth
import PIL
from PIL import Image, ImageTk 
from tkinter import *
from tkinter import ttk

class dashboard():
    
    brief_list = [] 
    model_list = []
    inputs_list =[]
    name_list =  []
    test_data_list = []
    # current_inps_lbs = []
    
    #model importing functions
    def modelimport(self,model,name,quote,inputs,testdata):
        self.brief_list.append(quote)
        self.model_list.append(model)
        self.name_list.append(name)
        self.test_data_list.append(testdata)
        self.menubutton(self.lf,name,model)
        self.inputs_list.append(inputs)
        return


    def mclick(self,x):
        global mid
        self.mindex = self.model_list.index(x)
        self.mid = self.mindex
        self.mfmtext.config(text = 'clicked')#self.brief_list[mindex]
        self.mfh.config(text = self.name_list[mindex])
    #     tdata_display(test_data_list[mindex])
        self.tdata.delete('1.0', END)
        self.tdata.insert(END, 'Test Data\n' + str(self.test_data_list[mindex]))
        self.inputs_display(self.inputs_list[mindex])
        return


    cinputs ={}
    cinplbls =[]
    def inputs_display(self,li):
    #     rf.pack_forget()
    #     rf.destroy()
        self.clear_inps()
        for i in li:
            lbl,ent = self.inputent(rf,i)
            self.cinputs[i[0]] = ent
            self.cinplbls.append(lbl)
        return

    #running the test
    def runtest(mid):
        return

    #result view
    def bfdisplay(data,imgs,root):
        for i in imgs:
            bimg = Label(root, image = i)
            bimg.pack()
        rtdata = Text(root, height=12, width=30)
        rtdata.insert(END, data)
        rtdata.pack()
        return


    #UI element Functions
    def fhead(self,frame,tex):
        heading = Label(frame,text=tex, font='Helvetica 12 bold')
        heading.pack(fill='x')
        return heading

    def menubutton(self,frame,tex,model):
        mbu = Button(frame, text=tex,command=mclick(model))#lambda :
        mbu.pack(fill='x')
        return mbu





    def tdata_display(self,frame):
    #     S = Scrollbar(mf)

        tdata = Text(frame, height=12, width=30)
    #     tdata.insert(END,'\nTest Data\n', 'big')
        tdata.insert(END, 'Select any model from menu')
        tdata.pack(fill='x')
    #     S.config(command=testdata.yview)
    #     testdata.config(yscrollcommand=S.set)
        return tdata





    def inputent(self,frame,inp):
        lbl = Label(frame,text=inp[0])
        lbl.pack()
        if len(inp)<2:  
            ent = Entry(frame)
            ent.pack(side=TOP,fill='x')
        else:
            ent = combox(frame,inp[1])
            ent.pack(side=TOP,fill='x')
        return lbl,ent
    def combox(self,window,li):
        combo = ttk.Combobox(window)
        combo['values']= list(li)
        combo.current(1) #set the selected item
        return combo
    def clear_inps(self):
        for i in list(self.cinputs.keys()):
            self.cinputs[i].destroy()
        for i in self.cinplbls:
            i.destroy()
    #     mbu.destroy()
        return





    ###helping functions
    def resimg(path):
        image = PIL.Image.open(path)
        photo = ImageTk.PhotoImage(image)
        return photo






    def close(self):
        self.main.mainloop()
        return

    ###demo



    def __init__(self):
        self.main = Tk()
        self.main.title('Dashboard')
        self.main.geometry('1000x800')
        self.mid = 0 
    
        rows = 0
        while rows < 100:
            self.main.rowconfigure(rows, weight=1)
            self.main.columnconfigure(rows, weight=1)
            rows += 1
    
        ##Main Frame
    
        self.notebook = ttk.Notebook(self.main)
        self.demo = Frame(self.notebook)
        self.arc = Frame(self.notebook)
        self.notebook.add(self.demo, text='Demo')
        self.notebook.add(self.arc, text='Architecture')
        self.notebook.grid(row=1, column=0, columnspan=100, rowspan=50, sticky='NESW')

        #left frame
        self.lf = Frame(self.demo, highlightbackground="black", highlightcolor="black",width=240, height=500, highlightthickness=1,bd= 0)
        self.lf.grid(row=0, column=0, rowspan=30,columnspan=20,sticky="nsew")
        self.lfh = self.fhead(self.lf,'Menu')
        # but1 = menubutton(lf,'Option1',model)







        #middle Frame
        self.mf = Frame(self.demo, highlightbackground="black", highlightcolor="black",width=520, height=500, highlightthickness=1,bd= 0)
        self.mf.grid(row=0, column=20,rowspan=30,columnspan=60,sticky="nsew")
        self.mfh = fhead(self.mf,'Controls')
        self.mfmtext = Label(self.mf,text="Select any model from menu")
        self.mfmtext.pack(fill='x')
        self.tdata = self.tdata_display(self.mf)
        # mbu = Button(mf, text="fhm",command=quit)
        # mbu.pack(side=TOP)



        self.runtestlbl = Label(self.mf,text="Run the test (need to fill the inputs if neccessary)")
        self.runtestlbl.pack(side = BOTTOM)
        self.testbutton = Button(self.mf, text="Run Test",command=runtest(mid))
        self.testbutton.pack(side=BOTTOM)



        #right Frame
        self.rf = Frame(self.demo, highlightbackground="black", highlightcolor="black",width=240, height=500, highlightthickness=1,bd= 0)
        self.rf.grid(row=0, column=80 ,rowspan=30,columnspan=20,sticky="nsew")
        self.rfh = self.fhead(self.rf,'Custom Inputs')
        # rlbl,rinp = inputent(rf,['test',[1,2,3,4]])
        # rlbl1,rip1 = inputent(rf,['another'])



        #stabilizing frame
        self.sft = Frame(self.demo, highlightbackground="black", highlightcolor="black", height=500, highlightthickness=1,bd= 0)
        self.sft.grid(row=0, column=100, rowspan=30,columnspan=1,sticky="nsew")
        self.sfb = Frame(self.demo, highlightbackground="black", highlightcolor="black", height=500, highlightthickness=1,bd= 0)
        self.sfb.grid(row=30, column=100, rowspan=20,columnspan=1,sticky="nsew")
        self.sfbv = Frame(self.demo, highlightbackground="black", highlightcolor="black",width=1000, highlightthickness=1,bd= 0)
        self.sfbv.grid(row=30, column=0,columnspan=100,sticky="nsew")


        #Results Frame
        self.bf = Frame(self.demo, highlightbackground="black", highlightcolor="black",width=1000, height=300, highlightthickness=1,bd= 0)
        self.bf.grid(row=30, column=0,rowspan=20,columnspan=100,sticky="nsew")
        self.bfh = self.fhead(self.bf,'Results')
        
    





    


    # test_img = resimg('test.png')
    # clear_inps()
    # bfdisplay('test data on bottom frame for demo',[img1,img2],bf)
    # bimg = Label(bf, image = img1)
    # bimg.pack()
    # res_img_display = Label(bf,image = test_img)


```
